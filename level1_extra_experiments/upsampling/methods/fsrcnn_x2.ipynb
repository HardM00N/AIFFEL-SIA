```python
import os
import glob
import numpy
import cv2
from cv2 import dnn_superres
from PIL import Image

# 지정 경로의 모든 png 파일 가져옴
for filename in glob.glob('/content/*.png'):
    # im=Image.open(filename)
    im = cv2.imread(filename)
    
    # Create an SR object
    fsrcnn_sr3 = dnn_superres.DnnSuperResImpl_create()
   
    # Read the desired model
    path = "/content/FSRCNN_Tensorflow/models/FSRCNN_x2.pb"
    fsrcnn_sr3.readModel(path)
    
    # Set the desired model and scale to get correct pre- and post-processing
    fsrcnn_sr3.setModel("fsrcnn", 2)

    # Upscale the image
    fsrcnn_result = fsrcnn_sr3.upsample(im)
    
    # Split path & name
    file_path, file_name = os.path.split(filename)

    # Save the image
    cv2.imwrite('/content/'+file_name.replace(".png", "")+'_x2.png', fsrcnn_result)
    print('/content/'+file_name.replace(".png", "")+'_x2.png')
```
